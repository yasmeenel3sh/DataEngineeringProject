```python
# Draw kdeplot

def kdeplot(dataframe,column):
    sns.kdeplot(dataframe[column])
    plt.title('Dist. of '+column)
    plt.show()

def heatmap(dataframe,columns):
    plt.figure(figsize=(10,5))
    df = dataframe[columns].corr()
    sns.heatmap(df, annot=True, cmap="YlGnBu", vmin=-1, vmax=1)
```

## Q4- Are the state and addr state correlated?

```python
# Create the contingency table
def contingencyTable(df,col1,col2):
    return pd.crosstab(df[col1], df[col2])
```

```python
import scipy.stats as stats

contingency_table= contingencyTable(fintech_indexed,'Addr State','State')
# Perform the Chi-Square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
chi2, p, dof
```

### p value indicate the state and addr are highly dependant

## Detecting Missing Values 


```python
# Calculating sum of null entries and percentage of null entries
sum_null = fintech_indexed.isnull().sum()
perc_null_mean = fintech_indexed.isnull().mean()*100
print(sum_null.sort_values(ascending=False))
print('-'*40)
perc_null_mean.sort_values(ascending=False)
```

### 1-Missing data Not At Random (MNAR): Systematic missing values

- We can infer that **annual_inc_joint** is MNAR since it is filled in case there is a co-borrower but that doesn't exist in most of the cases,
so the information is missing since it doesn't exist.

- The results indicate that out of 1926 missing records in **emp_length**, 1917 are missing when the **emp_title** is also missing indicating that when a title is not given, the employment length is also not given, thus **emp_length** is **MNAR** 

```python
fintech_indexed['emp_title'].value_counts()
```

```python
fintech_indexed['emp_length'].value_counts()
```

```python
# Check the number of records where both 'emp_length' and 'emp_title' are null
simultaneous_nulls = fintech_indexed[(fintech_indexed['emp_length'].isnull()) & (fintech_indexed['emp_title'].isnull())]
simultaneous_nulls.shape[0]
```

### 2- Missing at Random, MAR:

- **emp_title** appears to be a missing at random , it might be due to the fact that some people do not prefer to disclose their jobs.

#### 3-Missing Completely at Random, MCAR:

- **int_rate** doesn't indicate any relation with the term or the grade of the load in terms of missing values, thus it is highly likely to be miss entered rather than being missing due to a relation with other values or behaviour.

- **description** is also missing completely at random since the missing values are less than 1%.
