## 1-int_rate
 it showed a positive correlation with the grade column so we will explore the medians and means per grade, grade is a cetegorical value and int_rate is a skewed distribution. Therefore, it is better to impute it with the median value rather than the mean.

```python
fintech_indexed.groupby('grade')['int_rate'].agg(
  missing_values=lambda x: x.isnull().sum(),
  mean_rate='mean',
  median_rate='median'
)
```

```python
fintech_indexed['int_rate_imputed'] = fintech_indexed.groupby('grade')['int_rate'].transform(lambda x: x.fillna(x.median()))
```

```python

from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder



df = fintech_des.copy()

# Encode categorical columns
label_encoders = {}
for column in ['emp_title', 'emp_length', 'verification_status', 'home_ownership']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))
    label_encoders[column] = le



# Select relevant features for KNN imputation
features = ['emp_title', 'emp_length', 'annual_inc_noout', 'verification_status', 'home_ownership']

# Initialize KNNImputer
knn_imputer = KNNImputer(n_neighbors=200)

# Perform KNN imputation
df[features] = knn_imputer.fit_transform(df[features])

# Decode the imputed values back to original labels for 'emp_title' and 'emp_length'
df['emp_title'] = label_encoders['emp_title'].inverse_transform(df['emp_title'].round().astype(int))
df['emp_length'] = label_encoders['emp_length'].inverse_transform(df['emp_length'].round().astype(int))

df.groupby('verification_status')['emp_length'].agg(
  missing_values=lambda x: x.isnull().sum(),
)
```
